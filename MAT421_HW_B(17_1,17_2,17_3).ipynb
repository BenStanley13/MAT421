# 17.1 Interpolation Problem Statement


Given a data set [x_i,y_i], were the x data is independent and the y data is dependent. Interpolation can be used to create an estimation function, called an interpolation function, such that Y(x_i)=y_i.

This interpolation function can be used to estimate a y_i for a given x_i in the range of the orignal data set.


The interpolation function does not require an underlying model for data.


```python
import matplotlib.pyplot as plt
x=[1,2,3,5,6]
y=[1,4,3,6,7]
plt.figure
plt.plot(x,y,'go',4,5.5,'ro')
plt.title('Interpolation Data Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

# 17.2 Linear Interpolation

In linear interpolation the point that is being estimated is assumed to lie on the line connecting the points to the left and right of the point being estimated.

To show this point here is an example using the module SciPy function interp1d.
x = [1,2,3,5] y=[1,3,5,4]

```python
from scipy.interpolate import interp1d
plt.style.use('seaborn-poster')
```

```python
x=[1,2,3,5,6]
y=[1,4,3,6,7]
f = interp1d(x,y)
y_hat = f(4)
print(y_hat)
```

```python
plt.figure
plt.plot(x,y,'-og')
plt.plot(4,y_hat,'ro')
plt.title('Linear Interpolation at x = 4')
plt.ylabel('Y')
plt.xlabel('X')
plt.show()
```

# 17.3 Cubic Spline Interpolation

In cubic spline interpolation the interpolation function is a set of piecewise functions. The points (x_i,y_i) and (x_i+1,y_i+1) are assumed to be connected by a cubic polynomial S_i(x) = a_i*x^3 + b_i*x^2 + c_i*x + d_i .

Here is an example of Cubic Spline Interpolation for the data set x,y = ([1,2,3,4,5],[1,3,1,4,6]).

```python
from scipy.interpolate import CubicSpline
import numpy as np
plt.style.use('seaborn-poster')
```

```python
(x,y) = ([1,2,3,4,5],[1,3,1,4,6])
f = CubicSpline(x,y,bc_type='natural')
x_new = np.linspace(1,5,100)
y_new = f(x_new)
```

```python
plt.figure
plt.plot(x_new,y_new,'g')
plt.plot(x,y,'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('X')
plt.ylabel('Y')
```

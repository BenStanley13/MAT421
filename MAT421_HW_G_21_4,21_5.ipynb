## 21.4: Simpson"s Rule

Simpson’s Rule approximates the area under f(x)
 over these two subintervals by fitting a quadratic polynomial through the points (xi−1,f(xi−1)),(xi,f(xi))
, and (xi+1,f(xi+1))
, which is a unique polynomial, and then integrating the quadratic exactly.

construct the quadratic polynomial approximation of the function over the two subintervals. By applying the formula for constructing Lagrange polynomials we get the polynomial

Pi(x)=f(xi−1)(x−xi)(x−xi+1)(xi−1−xi)(xi−1−xi+1)+f(xi)(x−xi−1)(x−xi+1)(xi−xi−1)(xi−xi+1)+f(xi+1)(x−xi−1)(x−xi)(xi+1−xi−1)(xi+1−xi),

and with substitutions for h
 results in

Pi(x)=f(xi−1)2h2(x−xi)(x−xi+1)−f(xi)h2(x−xi−1)(x−xi+1)+f(xi+1)2h2(x−xi−1)(x−xi).

To approximate the integral over (a,b)
, we must sum the integrals of Pi(x)
 over every two subintervals since Pi(x)
 spans two subintervals. Substituting h3(f(xi−1)+4f(xi)+f(xi+1))
 for the integral of Pi(x).

```python
import numpy as np

a = 0
b = 6
n = 15
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.sin(x)

I_simp = (h/3) * (f[0] + 2*sum(f[:n-2:2]) \
            + 4*sum(f[1:n-1:2]) + f[n-1])
err_simp = 2 - I_simp

print(I_simp)
print(err_simp)
```

```python
import numpy as np
from scipy.integrate import trapz

a = 0
b = 5
n = 13
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.sin(x)

I_trapz = trapz(f,x)
I_trap = (h/2)*(f[0] + 2 * sum(f[1:n-1]) + f[n-1])

print(I_trapz)
print(I_trap)
```

```python
from scipy.integrate import cumtrapz
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('seaborn-poster')

x = np.arange(0, 2*np.pi, 0.015)
F_exact = -np.cos(x)
F_approx = cumtrapz(np.sin(x), x)

plt.figure(figsize = (10,6))
plt.plot(x, F_exact)
plt.plot(x[1::], F_approx)
plt.grid()
plt.tight_layout()
plt.title('$F(x) = \int_0^{x} sin(y) dy$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['Exact with Offset', 'Approx'])
plt.show()
```
